## Interpretation

**Key questions to consider:**

1. **Coefficient stability:** How does the IV coefficient change across models?
2. **Significance (statistical):** Is the IV statistically significant in all models?
3. **Sign (direction) of effect:** Is the coefficient positive or negative? What does this mean for our hypothesis?
4. **Strength of effect:** How does the coefficient of the IV compare to the other coefficients ? Is this a strong or small effect when we put it in the context of the hypothesis?
4. **Confounding:** Do the demographic controls substantially change the IV coefficient? What does this suggest?
5. **Model fit:** How does R-squared change as we add controls?
