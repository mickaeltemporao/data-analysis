```python
# Load library
import pandas as pd
```

```python
# Load the 2024 ANES example dataset
data_url = 'https://raw.githubusercontent.com/datamisc/ts-2024/main/data.csv'
df = pd.read_csv(data_url, compression='gzip')
df.head()
```

## Partisanship (V241043) – Bar Chart

```python
var = "V241043"
df[var].value_counts().plot(kind='bar', title='Partisanship')
```

## Ideology (V241177) – Ordered Bar Chart

```python
var = "V241177"
df[var].value_counts().sort_index().plot(kind='bar', title='Ideology (7-Point Scale)')
```

## Feeling Thermometer: Donald Trump (V241157) – Histogram

```python
var = "V241157"
df[var].plot(kind='hist', bins=30, title='Feeling Thermometer: Donald Trump')
```
